```python
import tkinter
from tkinter import *
import random
from tkinter import messagebox
```

```python

#  ########
words = [       #gusseing word in array
    "nptoyh",
    "avja",
    "wfsit",
    "cdanaa",
    "aidin",
    "aiearcm",
    "odnlon",
    "iscjavptia",
    "tmhl",
    "s3cs",
]


answers = [   #Currect answer for gusing
    "python",
    "java",
    "swift",
    "canada",
    "india",
    "america",
    "london",
    "javascript",
    "html",
    "css3",
]
```

```python
# random word   
num =  random.randrange(0, len(words), 1)

def default():  #defult show
    global words,answers,num
    lbl.config(text = words[num])

# resett button command
def res():
    global words,answers,num
    num = random.randrange(0, len(words), 1)
    lbl.config(text = words[num])
    e1.delete(0, END)


# check answer command
def checkans():
    global words,answers,num
    var = e1.get()
    if var == answers[num]:
        messagebox.showinfo("Success", "This is a correct answer")  #massagebox sent
        res()
    else:
        messagebox.showerror("Error", "This is not Correct")
        e1.delete(0, END)


root = tkinter.Tk()  #gui initiat
root.geometry("800x640")  #fram size
root.title("Guessing Game") #titile window
root.configure(background = "#34495e") #bg color
```

```python
# heading
lbl_text = Label(
    pady=20, text="Guess the word bellow", #heading titile
    font=('Roboto', 18),  #font initiat
    width=700,  
    bg='#34495e', #background color
    fg='#FFFFFF'  #text time
)

lbl_text.pack(fill=X) 


lbl = Label(  
    root,
    text = "",
    font = ("Roboto", 18),
    bg = "#3498db",
    fg = "#FFFFFF",
)
lbl.pack(pady = 30,ipady=10,ipadx=30) #padding teautorial


ans1 = StringVar()
e1 = Entry(  #answer
    root,
    font = ("Roboto", 16),
    textvariable = ans1,
)
e1.pack(ipady=5,ipadx=5)
```

```python
# check button
btncheck = Button( 
    root,
    text = "Check",
    font = ("Roboto", 16),
    width = 16,
    bg = "#27ae60",
    fg = "#FFFFFF",
    command = checkans, #checkans fuction call
    
)
btncheck.pack(pady = 50, ipady=5)


#  reset btton
btnreset = Button(
    root,
    text = "Reset",
    font = ("Roboto", 16),
    width = 16,
    bg = "#e74c3c",
    fg = "#FFFFFF",
    command = res, #res function
)
btnreset.pack(pady = 1, ipady=5)


default() # default
root.mainloop() 
```
